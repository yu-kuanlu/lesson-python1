```python
import random

def get_scores(names, num=10):
    stu_names = random.sample(names,num)
    scores = []
    for name in stu_names:
        info = {"姓名":name,
                "國文":random.randint(50, 100),
                "英文":random.randint(50, 100),
                "數學":random.randint(50, 100)
                }
        scores.append(info)
    return scores

students = get_scores(names,num=5)
students
```

```python
import csv

fieldnames = students[0].keys()
current_dir = os.path.dirname(os.path.abspath("__file__"))
file_path = os.path.join(current_dir,'assets','students.csv')

with open(file_path,'w',newline='',encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for d in students:
        writer.writerow(d)
```

```python
import csv

def save_csv(students,filename):
    fieldnames = students[0].keys()
    current_dir = os.path.dirname(os.path.abspath("__file__"))
    file_path = os.path.join(current_dir,'assets',filename)

    with open(file_path, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for d in students:
            writer.writerow(d)

save_csv(students,'students.csv')
```
