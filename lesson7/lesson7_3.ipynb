```python
cities = ['台北','桃園','新竹','苗栗']
cities
```

```python
# 索引index
cities[-1]
```

```python
# slice[前面的索引:後面的索引]
# 前面的所以包含,後面的索引不包含
print(cities[0:1])
print(cities[0:2])
print(cities[:2])
print(cities[:3])
print(cities[2:3])
print(cities[2:4])
print(cities[2:])
```

```python
greetings = list('hello lu')
print(greetings[:2])
print(greetings[3:])
print(greetings[2:5])
print(greetings[0])
print(greetings[6])
```
