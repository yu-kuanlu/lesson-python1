```python
# 處理錯誤
#應用程式增加可靠度

try:
    money = int(input("請輸入額度:"))
    if money < 0:
        raise Exception("額度不小於0")
except ValueError:
    print("錯誤錯誤")
except Exception as error:
    print(error)
else:
    print(money)
finally:
    print("程式結束")
```
